```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(100)
y=0.2*x+0.5
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.show()
```

```python
def plot_prediction(pred,y):
    plt.scatter(x,y)
    plt.scatter(x,pred)
    plt.show()
```

```python
w=np.random.uniform(-1,1)
b=np.random.uniform(-1,1)

LR=0.7

for epoch in range(100):
    y_pred=w*x+b
    error=np.abs(y_pred-y).mean()
    if error<0.0001:
        break
    w_grd=LR*((y_pred-y)*x).mean()
    b_grd=LR*(y_pred-y).mean()
    w=w-w_grd
    b=b-b_grd
    if epoch%5==0:
        y_pred=w*x+b
        plot_prediction(y_pred, y)
```
